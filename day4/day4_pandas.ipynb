```python
import pandas as pd
import numpy as np

data = {
    "Name" :["Alice","Bob","Charlie","David"],
    "Age" : [25,30,np.nan,22],
    "City" : ["pune","chennai","hyd","ban"],
    "Department" : ["HR","IT",np.nan,"HR"],
    "Salary" : [50000,60000,55000,  np.nan]
}
df = pd.DataFrame(data)

# Forward fill
df_ffill = df.fillna(method='ffill')
print("Forward Fill:\n", df_ffill)

# Backward fill
df_bfill = df.fillna(method='bfill')
print("\nBackward Fill:\n", df_bfill)
```

```python
sales = {
    "date" : ["2023-07-01","2023-07-02","2023-07-03","2023-07-04","2023-07-05"],
    "region" : ["North","South",np.nan,"West","North"],
    "product" : ["lap","phone","lap","earphone","phone"],
    "units_sold":[10,np.nan,15,12,15],
    "unit_price":[300,250,400,np.nan,150]
}

df = pd.DataFrame(sales)

print(df.isnull().sum())
print(df)
df.fillna({"unit_price": df["unit_price"].mean()}, inplace = True)
df.fillna({"units_sold": df["units_sold"].mean()}, inplace = True)
df.dropna(subset=["region"],inplace = True)
sold = df[df['units_sold'] > 10]
west_region = df[df['region'] == "West"]
print(west_region)
print(sold)
print(df)
print(df.groupby("region" )["units_sold"].max())
print(df[df["product"] == "phone"]["units_sold"].sum())
```
