```python
import matplotlib.pyplot as plt
marks = [10,20,30,40,50,60,70]

plt.boxplot(marks)
plt.grid()
plt.title("")
plt.show()
```

```python
from collections.abc import ValuesView
hours = [1,2,3,4,5,6]
scores = [40,50,65,86,54,89]
plt.scatter(hours,scores,color = 'green')

plt.grid()
plt.title("scores vs study hours")
plt.xlabel("study hours")
plt.ylabel("scores")

plt.show()
```

```python
import matplotlib.pyplot as plt
days = ['mon','tue','wed','thu','fri']
attendance = [90,80,78,78,60]

plt.plot(days,attendance , marker = 'o',linestyle = '--', color = 'Blue')
plt.title("Attendance in week 1 ")
plt.xlabel("Days")
plt.ylabel("Attendance")
plt.grid(True)
plt.show()
```

```python
subjects = ["python", "java","dsa","maths","physics"]
marks = [45,12,35,41,26]
plt.bar(subjects,marks)
plt.title("Marks in subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
activity = ["sleep","running","screentime","coding","cleaning"]
hours = [10,1,5,2,1]
plt.pie(hours,labels = activity,autopct='%1.1f%%',startangle = 90)
plt.title("Hours spent in different activities")
plt.legend(bbox_to_anchor = (1,1),loc = 2)
plt.show()
```

```python
ages = [10,20,40,25,36,89,45,25,53,54,45,26,28,29]
plt.hist(ages,bins = 6,color = "lightyellow",edgecolor = "black")
plt.title("Histogram of ages")
plt.xlabel("Ages")
plt.ylabel("Frequency")
plt.show()
```
